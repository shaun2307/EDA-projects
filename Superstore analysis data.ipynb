```python
import pandas as pd
data = pd.read_excel("Superstore data copy.xlsx")
print(data)
```

```python
data.head(10)
```

```python
duplicate_rows_data = data[data.duplicated()]
print("number of duplicate rows: ", duplicate_rows_data.shape)
```

```python
print(data.isnull().sum())
```

```python
data = data.drop(["Product Base Margin"], axis=1)
data.head(5)
```

```python
import seaborn as sns
sns.boxplot(x=data['Profit'])
```

```python
sns.boxplot(x=data['Sales'])
```

```python
import matplotlib.pyplot as plt
x= data["Product Category"]
y= data["Sales"]
plt.barh(x,y,color="b")
plt.title("Product Category Wise Sales")
plt.ylabel('Sales')
plt.xlabel('Product Category')
```

```python
x= data["Customer Segment"]
y= data["Sales"]

plt.bar(x,y,color="k",alpha=1)

plt.title("Customer Segment Sales")
plt.ylabel('Sales')
plt.xlabel('Customer Segment')
plt.show()
```

```python
x= data["Region"]
y= data["Sales"]
plt.scatter(x,y,color ="c")
plt.colorbar()
plt.title("Customer Segment Sales")
plt.ylabel('Sales')
plt.xlabel('Customer Segment')
plt.show()
```

```python
n = data["Region"]

plt.hist(n,orientation="horizontal", color="g", label="Regions")
plt.legend()
plt.grid()
plt.show()
```
