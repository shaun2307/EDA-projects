```python
import pandas as pd
data= pd.read_csv("ipl.csv")
print(data)
```

```python
data.info()
```

```python
data["player_of_the_match"]= str(data["player_of_the_match"])
```

```python
d1= data.dropna()
```

```python
data.head(5)
data.sample(5)
```

```python
import matplotlib.pyplot as plt
x = data["top_scorer"]
y = data["highscore"]
plt.barh(x,y,color="b",edgecolor="r",label="Highscore")
plt.legend()
plt.title("Batsmen with Highest Score")
plt.yticks(fontsize=6)
plt.show()
```

```python
d= data["match_winner"]
plt.hist(d,color="g",label="Match Winners",cumulative=-1,histtype="step")
plt.legend()
plt.xticks(fontsize=7)
plt.grid()
plt.title("Top Match Winners")
plt.show()
```

```python
x = data["match_winner"]
y = data["margin"]
plt.bar(x,y,color="b",edgecolor="y")
plt.title("Match Margin")

plt.xticks(fontsize=6)
plt.show()
```

```python
x = data["match_winner"]
y = data["won_by"]
s= [180]
plt.scatter(x,y,color="m",edgecolor="y",marker="*",sizes=s)
plt.title("Won By")
plt.grid()
plt.colorbar()
plt.xticks(fontsize=6)
plt.show()
```

```python
g = data["top_scorer"]
l = data["highscore"]

plt.step(g,l,color="m")
plt.title("Won By")
plt.grid()

plt.xticks(fontsize=2)
plt.show()
```
